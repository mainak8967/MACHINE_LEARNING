```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Plotting a vector

```python
plt.quiver(0,0,2,3)# helps in plotting vector connecting (0,0) to (2,3)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units="xy",angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units="xy",angles='xy',scale=1,color='g')
plt.quiver(0,0,-3,-4,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of Two Vectors

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum=vector_1+vector_2
print (sum)
```

```python
plt.quiver(0,0,2,3,scale_units="xy",angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of Two Vectors

```python
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
diff=vector_1-vector_2
print (diff)
```

```python
plt.quiver(0,0,2,3,scale_units="xy",angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot Product of Two Vectors

```python
a=np.asarray([2,3])# 2i cap+ 3 jcap
b=np.asarray([4,4])
a_dot_b=np.dot(a,b)
a_dot_b
```

```python
c=np.asarray([2,3,4])
d=np.asarray([4,5,6])
print(np.dot(c,d))
```

Cross Product of Two Vectors

```python
a=np.asarray([2,3,0])# if u wrote 2,3 then it would be equivalent to 2,3,0
b=np.asarray([4,4,0])#dimension must be 2 or 3
a_cross_b=np.cross(a,b)
a_cross_b
```

Projection Vector of a vector on another vector

```python
a=np.array([2,5])
v=np.array([8,-6])
#we will find projection of a on v so we will need magnitude of v
mod_v=np.sqrt(np.dot(v,v))
proj_of_a_on_v=((np.dot(a,v))/mod_v**2)*v
print(proj_of_a_on_v)
```
